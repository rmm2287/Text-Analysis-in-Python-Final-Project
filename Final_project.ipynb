# Method

When evaluating the 311 Call Center Inquiry data, it represents fairly basic call center information. The data collects the date and time of the calls, the agency most associated with the call, an inquiry name, a brief description, and the resolution of the call. For my project, I was able to utilize the date/time category to narrow down to the March to May time period I had decided on. Other than that initial search, only the inquiry name and the brief description were relevant to my research.

The first major challenge I experienced is that the New York Open Data platform will only allow for 1000 rows to be exported at a time. I was able to utilize ChatGPT to navigate this issue, and had to utilize that code over and over again to ensure the correct data was being analyzed throughout my data.

I performed the following methodology and analysis on the dataset. I wrote and ran the python script in a GoogleCollab notebook. Besides built-in python features, I used libraries including pandas, WordCloud, nltk, mathplotlib, textblob, VADER, and tomotopy. The methodology consisted of text processing and keyword analysis and theme categorization and sentiment analysis.

Within the text processing and keyword analysis I removed common stop words. I also added a list of other stopwords to exclude from the data set. These words included new, york, New York, complaint, repair, report, service, requested, needed, and issue. These were words that occurred often in the data set, and words in which the meaning and presence in the results were not generating any value to understanding the dataset. The test in the dataset was tokenized using the NLTK library which split the individual texts into words or tokes. In addition, all text was converted to lowercase and non-alphabetical characters like numbers were removed. Missing or incomplete entries were also removed from the inquiry name and brief description columns. The reprocessed data was then stores in processed_inquiry_name and processed_description, both representing data from the previously mentioned relevant columns of data.

Using this newly processed data, a WordCloud was created to show some of the most used words throughout the dataset. Some of the largest (and therefore, most common) terms in the WordCloud included camera violation, payment dispute, maintenance problem, and parking ticket. Then, the top twenty keywords were then pulled from the dataset along with their frequency of use.

To better understand the themes in the dataset, tomotopy was used which specified five topics for analysis based on the content within the dataset. Each document was represented as a sequence of words with the combination of inquiry name and brief description for each record being used as the input text. This data was then fed into the model and processed. The top words for each topic were extracted which allowed for and interpretation of the content. These top words provided insight into the key themes and issues prevalent in the dataset, which were essential for understanding the nature of the proposed research question. The topic assignments were then utilized to categorize the other terms in the dataset. This assignment process helped to structure the dataset into distinct thematic groups based on the content of the inquiries and descriptions.

Lastly, sentiment analysis was conducted to determine the emotional tone of the text. Using TextBlob and VADER, the polarity score of the text was calculated. A polarity score greater than 0 indicates a positive sentiment, less than 0 indicates a negative sentiment, and a score of signifies neutrality. For each word, the polarity was computed and classified. VADER in particular is often used for social media, but in this case, I chose to use it as it can be valuable in cases of service requests. VADER assigns compound scores which range from -1 (most negative) to 1 (most positive). Both sentiment analysis results were compared to provide a more comprehensive understanding of the emotional tone of the call center data inquiries.

```python
import requests
import pandas as pd
from io import StringIO
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')

url = 'https://data.cityofnewyork.us/resource/wewp-mm3p.csv'

limit = 1000
offset = 0

all_data = []
start_date = '2024-03-01'
end_date = '2024-05-31'

stop_words = set(stopwords.words('english'))
additional_stopwords = {'new york', 'new', 'york', 'report', 'complaint', 'repair', 'service', 'requested', 'needed', 'issue', 'get'}
stop_words.update(additional_stopwords)

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    return [word for word in tokens if word.isalpha() and word not in stop_words]

while True:
    params = {
        '$limit': limit,
        '$offset': offset,
        '$where': f"date >= '{start_date}' AND date <= '{end_date}'"
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        chunk_data = pd.read_csv(StringIO(response.text))
        all_data.append(chunk_data)

        if len(chunk_data) < limit:
            print(f"Fetched {len(chunk_data)} rows, stopping pagination.")
            break

        offset += limit
        print(f"Fetched {len(chunk_data)} rows, moving to offset {offset}.")
    else:
        print(f"Error {response.status_code}: Unable to fetch data.")
        break

final_data = pd.concat(all_data, ignore_index=True)

final_data['processed_description'] = final_data['brief_description'].apply(lambda x: preprocess_text(str(x)))

all_words = [word for desc in final_data['processed_description'] for word in desc]

word_counts = Counter(all_words)

top_20_words = word_counts.most_common(20)

top_20_df = pd.DataFrame(top_20_words, columns=['Word', 'Frequency'])
print(top_20_df)

plt.figure(figsize=(12, 8))
plt.barh(top_20_df['Word'], top_20_df['Frequency'], color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Top 20 Most Frequent Words')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

nltk.download('stopwords')

additional_stopwords = ['new york', 'new', 'york', 'report', 'complaint', 'repair', 'service', 'requested', 'needed', 'issue']

stop_words = set(stopwords.words('english')) | set(additional_stopwords)

data = {
    'brief_description': [
        'Pipe repair requested in New York.',
        'Water service complaint in York region.',
        'Complaint about new water pipe installation.',
        'Repair needed for broken pipe.',
        'Service issue with water pipe in New York.'
    ]
}

cleaned_data = pd.DataFrame(data)

def remove_stopwords(text):
    tokens = text.lower().split()

    return ' '.join([word for word in tokens if word not in stop_words])

cleaned_data['cleaned_description'] = cleaned_data['brief_description'].apply(remove_stopwords)

vectorizer = TfidfVectorizer(max_features=20)

X_matrix = vectorizer.fit_transform(cleaned_data['cleaned_description'])

terms = vectorizer.get_feature_names_out()

sum_tfidf = X_matrix.sum(axis=0).A1

term_scores = pd.DataFrame(zip(terms, sum_tfidf), columns=['Term', 'TF-IDF Score'])

top_20_terms = term_scores.sort_values(by='TF-IDF Score', ascending=False).head(20)

top_20_terms = top_20_terms[top_20_terms['Term'] != 'york']

print("\nTop 20 Terms based on TF-IDF (excluding 'york'):")
print(top_20_terms)

plt.figure(figsize=(12, 8))
plt.barh(top_20_terms['Term'], top_20_terms['TF-IDF Score'], color='skyblue')
plt.xlabel('TF-IDF Score')
plt.ylabel('Terms')
plt.title('Top Terms based on TF-IDF Scores (Excluding "york")')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd
from io import StringIO
import tomotopy as tp
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

url = 'https://data.cityofnewyork.us/resource/wewp-mm3p.csv'

limit = 1000
offset = 0

all_data = []

start_date = '2024-03-01'
end_date = '2024-05-31'

while True:
    params = {
        '$limit': limit,
        '$offset': offset,
        '$where': f"date >= '{start_date}' AND date <= '{end_date}'"
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        chunk_data = pd.read_csv(StringIO(response.text))
        all_data.append(chunk_data)

        if len(chunk_data) < limit:
            print(f"Fetched {len(chunk_data)} rows, stopping pagination.")
            break

        offset += limit
        print(f"Fetched {len(chunk_data)} rows, moving to offset {offset}.")
    else:
        print(f"Error {response.status_code}: Unable to fetch data.")
        break

final_data = pd.concat(all_data, ignore_index=True)

print(f"Total rows fetched within date range: {len(final_data)}")
print(final_data.head())

stop_words = set(stopwords.words('english'))
additional_stopwords = {'new york', 'new', 'york', 'report', 'complaint', 'repair', 'service', 'requested', 'needed', 'issue'}
stop_words.update(additional_stopwords)

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    return [word for word in tokens if word.isalpha() and word not in stop_words]

if 'inquiry_name' in final_data.columns:
    final_data['processed_inquiry_name'] = final_data['inquiry_name'].apply(lambda x: preprocess_text(str(x)))
else:
    final_data['processed_inquiry_name'] = []

if 'brief_description' in final_data.columns:
    final_data['processed_description'] = final_data['brief_description'].apply(lambda x: preprocess_text(str(x)))
else:
    final_data['processed_description'] = []

documents = final_data['processed_inquiry_name'].astype(str) + " " + final_data['processed_description'].astype(str)

model = tp.LDAModel(k=5)

for doc in documents:
    model.add_doc(doc.split())

model.train(1000)


print("\nTop words for each topic:")
for topic_id in range(model.k):
    print(f"\nTopic {topic_id + 1}:")
    words = model.get_topic_words(topic_id, top_n=10)
    for word, prob in words:
        print(f"{word}: {prob:.4f}")

print("\nTopic assignments for each document:")
for i, doc in enumerate(model.docs):
    topic_id = doc.topic
    print(f"Document {i + 1}: Topic {topic_id}")
```

```python
import requests
import pandas as pd
from io import StringIO
from textblob import TextBlob
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

url = 'https://data.cityofnewyork.us/resource/wewp-mm3p.csv'

limit = 1000
offset = 0

all_data = []

start_date = '2024-03-01'
end_date = '2024-05-31'

while True:
    params = {
        '$limit': limit,
        '$offset': offset,
        '$where': f"date >= '{start_date}' AND date <= '{end_date}'"
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        chunk_data = pd.read_csv(StringIO(response.text))

        all_data.append(chunk_data)

        if len(chunk_data) < limit:
            print(f"Fetched {len(chunk_data)} rows, stopping pagination.")
            break

        offset += limit
        print(f"Fetched {len(chunk_data)} rows, moving to offset {offset}.")
    else:
        print(f"Error {response.status_code}: Unable to fetch data.")
        break
final_data = pd.concat(all_data, ignore_index=True)

print(f"Total rows fetched within date range: {len(final_data)}")
print(final_data.head(50))

def analyze_sentiment_textblob(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        sentiment = 'Positive'
    elif sentiment_score < 0:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    return sentiment, sentiment_score

final_data['textblob_sentiment'], final_data['textblob_sentiment_score'] = zip(*final_data['brief_description'].apply(lambda x: analyze_sentiment_textblob(str(x))))

sia = SentimentIntensityAnalyzer()

def analyze_sentiment_vader(text):
    sentiment_score = sia.polarity_scores(text)
    compound_score = sentiment_score['compound']
    if compound_score > 0:
        sentiment = 'Positive'
    elif compound_score < 0:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    return sentiment, compound_score

final_data['vader_sentiment'], final_data['vader_sentiment_score'] = zip(*final_data['brief_description'].apply(lambda x: analyze_sentiment_vader(str(x))))

print("\nSentiment Analysis Results:")
print(final_data[['brief_description', 'textblob_sentiment', 'textblob_sentiment_score', 'vader_sentiment', 'vader_sentiment_score']].head(50))
```
